## 15与20, 区分日志中数据来源于1.5MW与2.0MW的不同表

```python
# add bwisfault is Y
for i in range(result_0.shape[0]):
    # 获取故障码对应的firstfault
    sql_hive_15 = "select windfarm,turbine,timestamputc,trigkeynum,firstfault from resmart."+ result_0['windfarm'][i]+ "_15 where turbine ="+result_0['turbine'][i]+\
    " and  firstfault ="+"'"+result_0['firstfault'][i]+"'"+" and trigkeynum regexp '^SC'"
#     print(sql_hive_15)
    hive_cursor.execute(sql_hive_15)
    result_15 = hive_cursor.fetchall()
    result_15 = pd.DataFrame(list(result_15))
    if result_15.empty:
        print(i,'empty',15)
    else:
        result_15.columns = ['windfarm','turbine','timestamputc','trigkeynum','firstfault']
        result_15['id'] = i
        pd.io.sql.to_sql(result_15,'pmc_slice_15',engine,if_exists='append')
        print(i,'success',15)


    sql_hive_20 = "select windfarm,turbine,timestamputc,trigkeynum,firstfault from resmart."+ result_0['windfarm'][i]+ "_20 where turbine ="+result_0['turbine'][i]+\
    " and  firstfault ="+"'"+result_0['firstfault'][i]+"'"+" and trigkeynum regexp '^SC'"
    hive_cursor.execute(sql_hive_20)
    result_20 = hive_cursor.fetchall()
    result_20 = pd.DataFrame(list(result_20))
    if result_20.empty:
        print(i,'empty',15)
    else:
        result_20.columns = ['windfarm','turbine','timestamputc','trigkeynum','firstfault']
        result_20['id'] = i
        pd.io.sql.to_sql(result_20,'pmc_slice_20',engine,if_exists='append')
        print(i,'success',20)
```
